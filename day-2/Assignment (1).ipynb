```python
l1 = [1,2,3,4]
l2 = [3,4,5,6]
result = []
for i in range(len(l1)):
    result.append(l1[i] + l2[i])
print(result)
```

```python
#Find maximum element combining two list l1 and l2
```

```python
combined_list = l1 + l2
max_element = max(combined_list)
print(max_element)
```

```python
d1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
d2 = {'a': 7, 'f': 2, 'g': 55, 'h': "101"}

#find common key from d1 and d2
#find common value from d1 and d2
```

```python
#find common key from d1 and d2
common_key = set(d1).intersection(d2)
print(common_key)
```

```python
#find common value from d1 and d2
common_value = set(d1.values()).intersection(d2.values())
print(common_value)
```

```python
#replace the values of d1 to their negatie value
replace_value = {key: -value for key, value in d1.items()}
print(replace_value)
```

```python
#replace key to value and value to key in d2
replace_key_value = {value: key for key, value in d2.items()}
print(replace_key_value)
```
